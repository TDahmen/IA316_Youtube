```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
x = np.linspace(0, 1, num=1000)
y1 = x * (1 - x)
y2 = y1 / 16

plt.plot(x, y1, "bo", label="x(1 - x)", markersize=1)
plt.plot(x, y2, "ro", label="x(1 - x) / 16", markersize=1)
plt.legend()
# plt.title("Variance = f(mean)")
plt.show()
```

```python
x = np.linspace(0, 1, num=1000)

m = 0.9
k = 16
a = (k - 1) * m
b = (k - 1) * (1 - m)
y1 = beta.pdf(x, a, b)

m = 0.6
k = 16
a = (k - 1) * m
b = (k - 1) * (1 - m)
y2 = beta.pdf(x, a, b)

plt.plot(x, y1, "go", label="Beta pdf, m = 0.9, k = 16", markersize=1)
plt.plot(x, y2, "co", label="Beta pdf, m = 0.6, k = 16", markersize=1)
plt.legend()
plt.show()
```
